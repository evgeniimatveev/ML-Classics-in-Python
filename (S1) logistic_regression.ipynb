# üîç Overview


Logistic Regression is one of the fundamental algorithms in **machine learning** used for **classification problems**. Unlike **Linear Regression**, which predicts continuous values, Logistic Regression estimates **the probability** that a given input belongs to a particular class (e.g., Spam or Not Spam üì©).

üí° **Why use Logistic Regression?**
- Simple and easy to interpret.
- Works well for linearly separable datasets.
- Outputs probabilities, allowing for decision-making with a threshold (e.g., 0.5).
- Serves as a baseline model before trying more complex classification algorithms like Random Forest or Neural Networks.

‚ö° **Common Applications of Logistic Regression:**
- **Email Classification** üìß (Spam vs. Not Spam)
- **Medical Diagnosis** üè• (Disease Present vs. Not Present)
- **Credit Scoring** üí≥ (Loan Default vs. No Default)
- **Marketing Campaigns** üìä (Will a customer buy the product?)

## üìå The Logistic Regression Formula:


$$
\hat{y} = \frac{1}{1 + e^{-(\beta_0 + \beta_1 x_1 + \beta_2 x_2 + ... + \beta_n x_n)}}
$$

### **Where:**
‚úî **\( \hat{y} \) (Predicted Probability)** ‚Üí The estimated probability that the target variable is **1** (e.g., will the customer buy? üõí)  
‚úî **\( \beta_0 \) (Intercept)** ‚Üí The bias term, shifting the decision boundary  
‚úî **\( \beta_1, \beta_2, ..., \beta_n \) (Coefficients)** ‚Üí The weights assigned to each feature  
‚úî **\( x_1, x_2, ..., x_n \)** ‚Üí Feature values used for prediction  
‚úî **\( e \) (Euler‚Äôs Number)** ‚Üí The base of the natural logarithm (~2.718)  

‚ö† **\( \epsilon \) (Error Term)** ‚Üí Represents model uncertainty and randomness in real-world data üîÑ  

---

## üîë **Key Insight:**
- **Logistic Regression** is a statistical model used for **binary classification** (e.g., spam vs. not spam üì©).  
- The **Sigmoid function** transforms predictions into probabilities between **0 and 1**.  
- Unlike **Linear R**e**gression**, Logistic Regression **does not predict continuous values** but rather the **likelihood** of an event.  
- **Decision Boundary** is determined by setting a probability threshold (e.g., 0.5).  
- Works well with **interpretable**, **linearly separable** problems, and **is the foundation** of many ML models!  

üí° **Logistic Regression is simple yet effective, making it a great choice for classification problems!** üöÄ

#üìå 1. Importing the Libraries

```
import numpy as np  # Numerical computations üî¢
import matplotlib.pyplot as plt  # Plotting graphs üìä
import pandas as pd  # Handling datasets üóÇÔ∏è
from sklearn.model_selection import train_test_split  # Splitting the dataset üîÄ
from sklearn.preprocessing import StandardScaler  # Feature scaling ‚öñÔ∏è
from sklearn.linear_model import LogisticRegression  # Logistic Regression Model üìà
from sklearn.svm import SVC  # Support Vector Machine üèÜ
from sklearn.ensemble import RandomForestClassifier  # Random Forest Classifier üå≤
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix  # Model evaluation metrics üìä
```

#üìå 2. Importing the Dataset

```
dataset = pd.read_csv('Social_Network_Ads.csv')  # Load dataset from CSV file üìÑ
X = dataset.iloc[:, :-1].values  # Select features (Age & Estimated Salary) üî¢
y = dataset.iloc[:, -1].values  # Select target variable (0 or 1) üéØ
```

# üìå 3. Splitting the Dataset

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)  # Split data into training (75%) & testing (25%) üîÄ
```

#üìå 4. Feature Scaling

```
sc = StandardScaler()  # Initialize Standard Scaler ‚öñÔ∏è
X_train = sc.fit_transform(X_train)  # Fit & transform training data üîÑ
X_test = sc.transform(X_test)  # Transform test data with the same scaler üîÑ
```

```
print(X_train)
```

```
print(X_test)
```

#üìå 5. Training Models

```
# Initialize Logistic Regression model with a fixed random state for reproducibility
logistic_model = LogisticRegression(random_state=0)
logistic_model.fit(X_train, y_train)  # Train Logistic Regression on the scaled training data üìà

# Initialize Support Vector Machine (SVM) model with a linear kernel
svm_model = SVC(kernel='linear', random_state=0)  # Linear kernel for SVM ensures a linear decision boundary üèÜ
svm_model.fit(X_train, y_train)  # Train the SVM model on the scaled training data

# Initialize Random Forest model with 100 decision trees (n_estimators=100)
rf_model = RandomForestClassifier(n_estimators=100, random_state=0)  # Ensemble learning with 100 trees üå≤
rf_model.fit(X_train, y_train)  # Train the Random Forest model on the scaled training data
```

#üìå 6. Predictions

```
y_pred_logistic = logistic_model.predict(X_test)  # Predict test set results using Logistic Regression üîÆ
y_pred_svm = svm_model.predict(X_test)  # Predict test set results using SVM üîÆ
y_pred_rf = rf_model.predict(X_test)  # Predict test set results using Random Forest üîÆ
```

#üìå 7. Evaluating Performance

```
metrics_data = {
    "Model": ["Logistic Regression", "SVM", "Random Forest"],  # Model names üèÜ
    "Accuracy": [
        accuracy_score(y_test, y_pred_logistic),  # Accuracy of Logistic Regression ‚úÖ
        accuracy_score(y_test, y_pred_svm),  # Accuracy of SVM ‚úÖ
        accuracy_score(y_test, y_pred_rf)  # Accuracy of Random Forest ‚úÖ
    ],
    "Precision": [
        precision_score(y_test, y_pred_logistic),  # Precision of Logistic Regression üéØ
        precision_score(y_test, y_pred_svm),  # Precision of SVM üéØ
        precision_score(y_test, y_pred_rf)  # Precision of Random Forest üéØ
    ],
    "Recall": [
        recall_score(y_test, y_pred_logistic),  # Recall of Logistic Regression üîç
        recall_score(y_test, y_pred_svm),  # Recall of SVM üîç
        recall_score(y_test, y_pred_rf)  # Recall of Random Forest üîç
    ],
    "F1-score": [
        f1_score(y_test, y_pred_logistic),  # F1-score of Logistic Regression üî¢
        f1_score(y_test, y_pred_svm),  # F1-score of SVM üî¢
        f1_score(y_test, y_pred_rf)  # F1-score of Random Forest üî¢
    ],
    "ROC-AUC": [
        roc_auc_score(y_test, y_pred_logistic),  # ROC-AUC score of Logistic Regression üìà
        roc_auc_score(y_test, y_pred_svm),  # ROC-AUC score of SVM üìà
        roc_auc_score(y_test, y_pred_rf)  # ROC-AUC score of Random Forest üìà
    ]
}
# Convert metrics data to DataFrame üìä
metrics_df = pd.DataFrame(metrics_data)
```

# üìå 8. Confusion Matrices


```

conf_matrix_logistic = confusion_matrix(y_test, y_pred_logistic)  # Confusion matrix for Logistic Regression üé≠
conf_matrix_svm = confusion_matrix(y_test, y_pred_svm)  # Confusion matrix for SVM üé≠
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)  # Confusion matrix for Random Forest üé≠
```

#üìå 9. Function to Plot Confusion Matrix

```
# üìå 9. Function to Plot Confusion Matrix
import seaborn as sns  # Import seaborn for visualization üé®

def plot_confusion_matrix(cm, model_name):
    plt.figure(figsize=(5, 4))  # Set figure size üìè
    class_labels = ["Class 0", "Class 1"]  # Define class labels

    sns.heatmap(cm, annot=True, fmt="g", cmap="Blues", xticklabels=class_labels, yticklabels=class_labels)  # Create heatmap üî•

    plt.xlabel("Predicted Label")  # X-axis label üìä
    plt.ylabel("True Label")  # Y-axis label üìä
    plt.title(f"Confusion Matrix - {model_name}")  # Title üìå
    plt.show()  # Display plot üìä
```

#  üìå 10. Visualizing Confusion Matrices


```
plot_confusion_matrix(conf_matrix_logistic, "Logistic Regression")  # Logistic Regression Matrix üé≠
plot_confusion_matrix(conf_matrix_svm, "SVM")  # SVM Matrix üé≠
plot_confusion_matrix(conf_matrix_rf, "Random Forest")  # Random Forest Matrix üé≠
```

#üìå  Visualizing Decision Boundaries (Updated and Integrated)

```
from matplotlib.colors import ListedColormap

# Function to plot decision boundaries for classification models
def plot_decision_boundary(model, title, X_set, y_set):
    """
    Plots the decision boundary of a classification model.

    Parameters:
    model  - The trained classification model.
    title  - Title for the plot.
    X_set  - Feature set for visualization.
    y_set  - True labels.
    """
    X1, X2 = np.meshgrid(np.arange(start=X_set[:, 0].min() - 1, stop=X_set[:, 0].max() + 1, step=0.01),
                         np.arange(start=X_set[:, 1].min() - 1, stop=X_set[:, 1].max() + 1, step=0.01))

    # Decision boundary visualization
    plt.contourf(X1, X2, model.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
                 alpha=0.75, cmap=ListedColormap(('red', 'green')))  # Red = Class 0, Green = Class 1

    plt.xlim(X1.min(), X1.max())  # Set X-axis limits
    plt.ylim(X2.min(), X2.max())  # Set Y-axis limits

    # Plot dataset points
    for i, j in enumerate(np.unique(y_set)):
        plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                    color=ListedColormap(('red', 'green'))(i), label=f"Class {j}")

    plt.title(title)  # Set title üèÜ
    plt.xlabel('Age')  # Label X-axis üéöÔ∏è
    plt.ylabel('Estimated Salary')  # Label Y-axis üéöÔ∏è
    plt.legend()  # Show legend üî•
    plt.show()  # Display plot üìä

# Ensure models are trained before calling the function
plot_decision_boundary(logistic_model, 'Logistic Regression (Test Set)', X_test, y_test)
plot_decision_boundary(svm_model, 'Support Vector Machine (Test Set)', X_test, y_test)
plot_decision_boundary(rf_model, 'Random Forest (Test Set)', X_test, y_test)
```
