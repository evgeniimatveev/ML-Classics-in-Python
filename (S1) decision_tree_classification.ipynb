# üî∑ Overview

Decision Tree (**DT**) is a popular and interpretable algorithm used for **classification tasks** based on a tree-like model of decisions. It splits data into subsets based on feature values, making it a **powerful yet explainable** model for various real-world applications. DT classifiers are widely used in **credit scoring, medical diagnosis, fraud detection, and recommendation systems**.

## ‚ùì What is Decision Tree?

Decision Tree classifiers use a **tree-based structure** to make predictions by recursively splitting data based on feature conditions. Each internal node represents a decision rule, each branch represents an outcome, and each leaf node represents a class label. **DT performs well** in many applications, especially with **structured and categorical data**.

## üî• Popular Decision Tree Variants:

- **ID3 (Iterative Dichotomiser 3)** üå± Uses entropy and information gain for feature selection.
- **C4.5** üå≥ An extension of ID3 that handles both categorical and continuous data.
- **CART (Classification and Regression Trees)** üìä Uses Gini impurity or entropy, supports both classification and regression.
- **CHAID (Chi-Square Automatic Interaction Detector)** üèÜ Uses chi-square statistics for multi-way splits.

---

üí° **Key Strengths**: Easy to interpret, no need for feature scaling, handles both numerical and categorical data.

‚ö†Ô∏è **Limitations**: Prone to overfitting, sensitive to noisy data, biased towards dominant features.

# üå≥ The Decision Tree Formula:

A Decision Tree (**DT**) model makes predictions by **recursively splitting** the dataset based on feature values until it reaches a class label.

**Mathematically, it minimizes impurity:**
$$
I(t) = - \sum_{i=1}^{c} p_i \log_2(p_i)
$$
where:
- \( I(t) \) is the impurity measure (e.g., Entropy or Gini).
- \( p_i \) is the probability of class \( i \) at node \( t \).

### Where:
‚úî **(I(t)) (Impurity Measure)** ‚Üí Measures how mixed the classes are at a given node. Common choices:
  - **Entropy:** Used in **ID3, C4.5**, based on information gain.
  - **Gini Impurity:** Used in **CART**, measures the probability of incorrect classification.

‚úî **(Split Criterion)** ‚Üí The method used to choose the best feature for splitting:
  - **Information Gain:** Used in ID3 and C4.5.
  - **Gini Index:** Used in CART.

‚úî **(Tree Nodes)** ‚Üí A decision tree consists of:
  - **Root Node:** The first decision point.
  - **Internal Nodes:** Where decisions are made.
  - **Leaf Nodes:** Represent class labels.

‚úî **(Recursive Splitting)** ‚Üí Process of dividing data into subsets until each subset belongs to a single class or meets a stopping condition.

---

## üîπ Choosing the Right Decision Tree Variant:

- **ID3 (Iterative Dichotomiser 3)** üå± Uses **Entropy & Information Gain** for selecting the best split.
- **C4.5** üå≥ An improvement over ID3 that handles **both categorical & numerical data**.
- **CART (Classification and Regression Trees)** üìä Uses **Gini impurity**, supports **both classification & regression**.
- **CHAID (Chi-Square Automatic Interaction Detector)** üèÜ Uses **chi-square statistics** for multi-way splits.

---

## üîë Key Insight:

‚úÖ **Handles both numerical and categorical data**, providing flexibility in applications.

# üìå Import Libraries

```
import numpy as np  # üìå Importing NumPy for numerical operations
import matplotlib.pyplot as plt  # üìå Importing Matplotlib for visualization
import pandas as pd  # üìå Importing Pandas for data manipulation
import seaborn as sns  # üìå Importing Seaborn for enhanced visualization
from sklearn.model_selection import train_test_split  # üìå Function to split dataset into training and testing sets
from sklearn.preprocessing import StandardScaler  # üìå Standardizing features for improved model performance
from sklearn.tree import DecisionTreeClassifier, plot_tree  # üìå Importing Decision Tree classifier and visualization tool
from sklearn.metrics import classification_report, confusion_matrix  # üìå Functions to evaluate model performance
from matplotlib.colors import ListedColormap  # üìå Used for coloring decision boundary plots
```

# üìå Load dataset from CSV file

```
# üìå Load dataset from CSV file
file_path = "Social_Network_Ads.csv"  # üìå File path for dataset
df = pd.read_csv(file_path)  # üìå Reading CSV into DataFrame
```

# üìå Display first few rows of dataset to understand structure

```
# üìå Display first few rows of dataset to understand structure
display(df.head())
```

# üìå Extracting features (Age, EstimatedSalary) and target variable (Purchased)

```
# üìå Extracting features (Age, EstimatedSalary) and target variable (Purchased)
X = df[['Age', 'EstimatedSalary']].values  # üìä Feature matrix: Age & Estimated Salary
y = df['Purchased'].values  # üéØ Target variable: Purchased (0 or 1)
```

# üìå Splitting data into training (75%) and testing (25%) sets

```
# üìå Splitting data into training (75%) and testing (25%) sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

# üìå Standardizing the features for better model performance



```
# üìå Standardizing the features for better model performance
sc = StandardScaler()
X_train = sc.fit_transform(X_train)  # üèãÔ∏è‚Äç‚ôÇÔ∏è Fitting and transforming training data
X_test = sc.transform(X_test)  # üîÑ Transforming test data using same scaler
```

# üìå Training a Decision Tree Classifier with a maximum depth of 4

```
# üìå Training a Decision Tree Classifier with a maximum depth of 4
classifier = DecisionTreeClassifier(criterion='entropy', random_state=42, max_depth=4)
classifier.fit(X_train, y_train)  # üöÄ Training the classifier
```

# üìå Function to plot decision boundaries

```
# üìå Function to plot decision boundaries
def plot_decision_boundary(X_set, y_set, model, title):
    X1, X2 = np.meshgrid(
        np.arange(start=X_set[:, 0].min() - 1, stop=X_set[:, 0].max() + 1, step=0.01),
        np.arange(start=X_set[:, 1].min() - 1, stop=X_set[:, 1].max() + 1, step=0.01)
    )  # üîπ Creating a grid for visualization
    plt.figure(figsize=(10, 6))  # üñº Setting figure size
    plt.contourf(X1, X2, model.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
                 alpha=0.5, cmap=ListedColormap(('lightblue', 'lightcoral')))  # üé® Plotting decision regions

    # üìå Scatter plot for actual data points
    for i, j in enumerate(np.unique(y_set)):
        plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                    c=ListedColormap(('blue', 'red'))(i), label=f'Class {j}')  # üìä Plotting data points

    plt.title(title)  # üè∑ Setting title
    plt.xlabel('Age (scaled)')  # üî¢ Label for X-axis
    plt.ylabel('Estimated Salary (scaled)')  # üí∞ Label for Y-axis
    plt.legend()  # üìú Adding legend
    plt.show()  # üñº Display plot
```


# üìå Plot decision boundaries for training data

```

# üìå Plot decision boundaries for training data
plot_decision_boundary(X_train, y_train, classifier, "Decision Tree Decision Boundary (Training Set)")
```

# üìå Plot decision boundaries for testing data

```

# üìå Plot decision boundaries for testing data
plot_decision_boundary(X_test, y_test, classifier, "Decision Tree Decision Boundary (Testing Set)")
```

# üìå Visualizing the decision tree structure

```
# üìå Visualizing the decision tree structure
plt.figure(figsize=(12, 8))  # üñº Setting figure size for tree visualization
plot_tree(classifier, feature_names=['Age', 'EstimatedSalary'], class_names=['No', 'Yes'], filled=True)  # üå≥ Plotting the tree
plt.title("Decision Tree Visualization")  # üè∑ Adding title
plt.show()  # üñº Display tree visualization
```

# üìå Testing different depths to analyze performance

```

# üìå Testing different depths to analyze performance
depths = [2, 4, 6, 8, 10]
for depth in depths:
    dt = DecisionTreeClassifier(max_depth=depth, random_state=42)  # üèó Creating new Decision Tree with different depths
    dt.fit(X_train, y_train)  # üöÄ Training the model
    print(f'üå≥ Max Depth {depth}: Accuracy = {dt.score(X_test, y_test):.4f}')  # üìä Printing accuracy for each depth
```

# üìå Generating predictions on test set

```
# üìå Generating predictions on test set
y_pred = classifier.predict(X_test)  # üéØ Predicting labels for test data
```

# üìå Confusion Matrix visualization

```
# üìå Confusion Matrix visualization
conf_matrix = confusion_matrix(y_test, y_pred)  # üìä Computing confusion matrix
plt.figure(figsize=(6, 5))  # üñº Setting figure size
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['No', 'Yes'], yticklabels=['No', 'Yes'])  # üé® Creating heatmap
plt.xlabel('Predicted')  # üî¢ X-axis label
plt.ylabel('Actual')  # üéØ Y-axis label
plt.title('Confusion Matrix')  # üè∑ Title
plt.show()  # üñº Display heatmap
```

# üìä Printing classification report

```
# üìå Printing classification report
print("üìä Classification Report:\n")  # üè∑ Title for report
print(classification_report(y_test, y_pred))  # üìú Printing precision, recall, F1-score
```
