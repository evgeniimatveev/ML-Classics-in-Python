## üìå Simple Linear Regression  

üìç **The Regression Formula:**  
# $\hat{y} = b_0 + b_1 X$
### **Where:**
‚úîÔ∏è **ùë¶ÃÇ (Predicted Value)** ‚Üí The estimated outcome (e.g., salary üí∞)  
‚úîÔ∏è **X (Independent Variable)** ‚Üí The input feature (e.g., years of experience üìÖ)  
‚úîÔ∏è **b‚ÇÄ (Intercept)** ‚Üí The starting value when **X = 0** (base salary) üèÅ  
‚úîÔ∏è **b‚ÇÅ (Slope Coefficient)** ‚Üí The rate at which **ùë¶** changes for each unit of **X** üìà  

---
**üí° Key Insight:**  
- The **intercept (b‚ÇÄ)** represents the salary when experience is **zero**.  
- The **slope (b‚ÇÅ)** shows how much the salary increases per additional year of experience.  
- If **b‚ÇÅ is positive**, there‚Äôs a positive correlation (more experience ‚Üí higher salary).  
- If **b‚ÇÅ is negative**, there‚Äôs a negative correlation (more experience ‚Üí lower salary).

```python
# üöÄ Simple Linear Regression - Comprehensive Script with Theory & Visualization
# ------------------------------------------------------------------------------
# üìå What is Linear Regression?
# Linear Regression is a fundamental machine learning algorithm used for predicting numerical values.
# It assumes a **linear relationship** between an independent variable (X) and a dependent variable (y).
# ------------------------------------------------------------------------------
# üîπ Formula for Simple Linear Regression:
#
#     ≈∑ = b‚ÇÄ + b‚ÇÅ * X
#
# Where:
# - ≈∑  ‚Üí Predicted value (dependent variable)
# - X  ‚Üí Independent variable (feature)
# - b‚ÇÄ ‚Üí Intercept (bias term)
# - b‚ÇÅ ‚Üí Slope (coefficient that determines the impact of X on y)
# ------------------------------------------------------------------------------
```

## Importing the libraries

```python
# üìå 1. Import Required Libraries
# ------------------------------------------------------------------------------
import numpy as np  # NumPy for mathematical operations and array handling
import matplotlib.pyplot as plt  # Matplotlib for plotting graphs
import pandas as pd  # Pandas for handling tabular data (dataframes)
```

## Importing the dataset

```python
# üìå 2. Load the Dataset
# ------------------------------------------------------------------------------
# The dataset "Salary_Data.csv" contains two columns:
# 1Ô∏è‚É£ "YearsExperience" ‚Üí Work experience in years (X)
# 2Ô∏è‚É£ "Salary" ‚Üí Salary in USD ($) (y)
dataset = pd.read_csv('Salary_Data.csv')
# Splitting dataset into independent (X) and dependent (y) variables
X = dataset.iloc[:, :-1].values  # Selecting all rows, but excluding the last column
y = dataset.iloc[:, -1].values  # Selecting the last column (Salary)
```

## Splitting the dataset into the Training set and Test set

```python
# üìå 3. Splitting the Data into Training & Test Sets
# ------------------------------------------------------------------------------
from sklearn.model_selection import train_test_split

# ‚úÖ 80% of data will be used for training
# ‚úÖ 20% of data will be used for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Training the Simple Linear Regression model on the Training set

```python
# üìå 4. Train the Linear Regression Model
# ------------------------------------------------------------------------------
from sklearn.linear_model import LinearRegression

# Initializing the Linear Regression model
regressor = LinearRegression()

# Training (fitting) the model on the training dataset
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

```python
# üìå 5. Predict the Test Set Results
# ------------------------------------------------------------------------------
# Predicting salaries based on the trained model
y_pred = regressor.predict(X_test)
```

## Visualising the Training set results

```python
# ------------------------------------------
# üé® Regression Visualization
# ------------------------------------------
import numpy as np
import matplotlib.pyplot as plt

# üìè Set figure size for a cinematic look
plt.figure(figsize=(12, 7))

# üî¥ Scatter plot for actual data points
plt.scatter(X_train, y_train, color='red', label='Actual Data', edgecolors='black', s=75, alpha=0.8)

# üîµ Adjusting regression line to be properly aligned
X_range = np.linspace(min(X_train), max(X_train), 100).reshape(-1, 1)  # Generate a smooth range
plt.plot(X_range, regressor.predict(X_range),
         color='blue', linewidth=3, linestyle='--',
         label='Regression Line')

# üé® Customizing the visualization for a professional feel
plt.title(' Salary vs Experience (Training Set)', fontsize=16, fontweight='bold', color='#333')
plt.xlabel(' Years of Experience', fontsize=14, fontweight='bold', color='#555')
plt.ylabel(' Salary ($)', fontsize=14, fontweight='bold', color='#555')

#  Enhanced Grid and Legend
plt.grid(True, linestyle='-.', linewidth=0.75, alpha=0.7)
plt.legend(facecolor='white', edgecolor='black', fontsize=12, loc='upper left', framealpha=1)

#  Aesthetic touch with a dark background
plt.style.use('fivethirtyeight')

#  Show the perfected visualization
plt.show()
```

## Visualising the Test set results

```python
# ------------------------------------------
# üéØ Enhanced Test Set Visualization
# ------------------------------------------
import numpy as np
import matplotlib.pyplot as plt

# üìè Set figure size for a polished look
plt.figure(figsize=(12, 7))

# üî¥ Scatter plot for actual test data points
plt.scatter(X_test, y_test, color='red', label='Actual Data', edgecolors='black', s=75, alpha=0.8)

# üîµ Regression Line (Ensuring Proper Alignment)
X_range = np.linspace(min(X_test), max(X_test), 100).reshape(-1, 1)  # Generate a smooth range
plt.plot(X_range, regressor.predict(X_range),
         color='blue', linewidth=3, linestyle='--',
         label='Regression Line')

# üé® Customizing the visualization
plt.title(' Salary vs Experience (Test Set)', fontsize=16, fontweight='bold', color='#333')
plt.xlabel(' Years of Experience', fontsize=14, fontweight='bold', color='#555')
plt.ylabel(' Salary ($)', fontsize=14, fontweight='bold', color='#555')

# üèÜ Grid and Legend for better readability
plt.grid(True, linestyle='-.', linewidth=0.75, alpha=0.7)
plt.legend(facecolor='white', edgecolor='black', fontsize=12, loc='upper left', framealpha=1)

# üåü Aesthetic touch with a modern theme
plt.style.use('fivethirtyeight')

# üöÄ Show the enhanced test set visualization
plt.show()
```

# Understanding Model Coefficients

```python
# üìå 7. Understanding Model Coefficients
# ------------------------------------------------------------------------------
# üîπ Once the linear regression model is trained, it calculates two key values:
#
#    1Ô∏è‚É£ Intercept (b‚ÇÄ) ‚Üí The starting point of the regression line on the y-axis.
#    2Ô∏è‚É£ Slope (b‚ÇÅ) ‚Üí The rate of change of y with respect to x (how much y changes when x increases by 1 unit).
#
# These values help us define the equation of the regression line:
#
#    ≈∑ = b‚ÇÄ + b‚ÇÅ * X
#
# Let's extract and display these coefficients:

b0 = regressor.intercept_  # Intercept (b‚ÇÄ) ‚Üí Base salary when experience is 0
b1 = regressor.coef_[0]    # Slope (b‚ÇÅ) ‚Üí How much salary increases per year of experience

# Printing the calculated coefficients with clear formatting
print(f"üìå üî• Intercept (b‚ÇÄ): {b0:.2f}")  # Displays base salary when experience = 0
print(f"üìå üî• Coefficient (b‚ÇÅ): {b1:.2f}")  # Displays increase in salary per extra year

# üîç **Example Interpretation**
# Let's say:
# - b‚ÇÄ (Intercept) = $26,7*80.10
# - b‚ÇÅ (Slope) = $9,312.58
# This means:
# - A person with **0 years of experience** is expected to earn **$26,780.10** (b‚ÇÄ).
# - For **each additional year** of experience, salary increases by **$9,312.58** (b‚ÇÅ).

# ------------------------------------------------------------------------------
```

# Validating the Regression Equation


```python
# ------------------------------------------
# üî• 8. Validating the Regression Equation
# ------------------------------------------
# Now, let's validate our regression model using the core equation of Linear Regression:
#
#   ≈∑ = b‚ÇÄ + b‚ÇÅ * X
#
# - ≈∑: Predicted salary
# - b‚ÇÄ: Intercept (base salary when experience = 0)
# - b‚ÇÅ: Slope (increase in salary per year of experience)
# - X: Years of experience
#
# The goal is to verify if our model correctly follows this equation.

# ------------------------------------------
# üéØ Step 1: Predict Salary for 12 Years of Experience
# ------------------------------------------
# - We choose **12 years of experience** as an example.
# - Convert it into a **2D array** for model input.
# - Pass it through our trained regression model.

years_of_experience = np.array([[12]])  # Convert to 2D array
predicted_salary = regressor.predict(years_of_experience)  # Predict salary

# ------------------------------------------
# üìå Step 2: Display the Predicted Salary
# ------------------------------------------
# - Print the predicted salary in a well-formatted way.
# - Adds **thousands separator and two decimal places** for clarity.

print(f"üî•üöÄ **Predicted Salary for 12 years of experience**: **${predicted_salary[0]:,.2f}**")

# ------------------------------------------
# üìä Step 3: Manually Verifying the Regression Formula
# ------------------------------------------
# To ensure correctness, we manually calculate the salary:
#
# Given:
#   - Intercept (b‚ÇÄ) = 26,780.10
#   - Slope (b‚ÇÅ) = 9,312.58
#   - Experience (X) = 12
#
# Manual Calculation:
#   ≈∑ = 26,780.10 + (9,312.58 √ó 12)
#   ≈∑ = 26,780.10 + 111,750.96
#   ≈∑ ‚âà **$138,531.00**
#
# ‚úÖ The manually calculated value **matches the model‚Äôs prediction**, confirming that our regression model is correctly following the equation.

# ------------------------------------------
# ‚úÖ Final Conclusion: Model Verification Complete!
# ------------------------------------------
# - The model correctly applies the linear regression equation.
# - The **predictions are accurate**, proving the model is functioning as expected.
# - We can confidently use this model for real-world predictions!
```

#  Making Predictions with New Data

```python
# ------------------------------------------
# üîÆ 9. Making Predictions with New Data
# ------------------------------------------
# Now that the model is trained, we can **predict salaries** for any given experience level.
# We can feed in new experience values to the model and get predicted salaries.

# ------------------------------------------
# üìå Step 1: Define New Experience Levels
# ------------------------------------------
# - Let's predict salaries for **different levels of experience**.
# - We create a NumPy array with various experience levels.
# - The model will **predict** salaries based on these inputs.

new_experience = np.array([[5], [8], [15], [20]])  # Define multiple experience levels

# ------------------------------------------
# üéØ Step 2: Use the Model to Predict Salaries
# ------------------------------------------
# - We pass our new experience data into the trained model.
# - The model will output the predicted salaries.

predicted_salaries = regressor.predict(new_experience)

# ------------------------------------------
# üìå Step 3: Print Predicted Results
# ------------------------------------------
# - We iterate through the predicted salaries and print them.
# - Each **experience level** is paired with its **predicted salary**.

for exp, salary in zip(new_experience.flatten(), predicted_salaries):
    print(f"üîπ Experience: {exp} years ‚Üí Predicted Salary: ${salary:,.2f}")

# ------------------------------------------
# üìä Observations & Real-World Use Cases
# ------------------------------------------
# ‚úÖ We can now use this model in **real-world applications**, such as:
#     ‚úî HR Salary Prediction Systems
#     ‚úî Career Growth Analysis
#     ‚úî Financial Forecasting
#     ‚úî AI-based Job Market Insights

# ------------------------------------------------------------------------------
```

# Evaluating Model Performance

```python
# ------------------------------------------
# üéØ 10. Evaluating Model Performance
# ------------------------------------------
# Now, we need to **evaluate how well our model performs**.
# We use key metrics like:
# - **Mean Absolute Error (MAE)**: Measures average absolute prediction error.
# - **R¬≤ Score**: Measures how well the model explains variance in the data.

from sklearn.metrics import mean_absolute_error, r2_score  # Importing evaluation metrics

# ------------------------------------------
# üèÜ Step 1: Calculate Performance Metrics
# ------------------------------------------

# üîπ Mean Absolute Error (MAE)
# -------------------------------------
# - Measures the **average absolute difference** between actual values (y_test) and predicted values (y_pred).
# - It **does NOT** consider direction (positive/negative), only absolute errors.
# - Formula: MAE = (1/n) * Œ£ | y_actual - y_predicted |
# - Lower MAE = **Better Model Performance**.

mae = mean_absolute_error(y_test, y_pred)

# üîπ R¬≤ Score (Coefficient of Determination)
# -------------------------------------
# - R¬≤ measures **how well the regression model fits the data**.
# - Formula: R¬≤ = 1 - (SS_residual / SS_total)
# - **Interpreting R¬≤:**
#     - **1.0** ‚Üí Perfect fit (explains 100% variance).
#     - **‚â• 0.8** ‚Üí Strong fit (explains ~80% of variance).
#     - **0.5 - 0.8** ‚Üí Moderate fit (explains ~50-80% variance).
#     - **< 0.5** ‚Üí Weak fit (model doesn‚Äôt explain much variance).

r2 = r2_score(y_test, y_pred)

# ------------------------------------------
# üñ® Step 2: Print the Evaluation Metrics
# ------------------------------------------
print(f"üìä Mean Absolute Error (MAE): {mae:.2f}")
print(f"üìà R¬≤ Score: {r2:.4f} (closer to 1 = better fit)")

# ------------------------------------------
# üìå Understanding R¬≤ Score - Quick Guide
# ------------------------------------------
# - **If R¬≤ = 1**, the model **perfectly fits** the data.
# - **If R¬≤ = 0.8**, the model explains **80%** of the variance.
# - **If R¬≤ < 0.5**, the model is **not a good fit**.

# ------------------------------------------
# ‚úÖ Summary of Model Evaluation
# ------------------------------------------
# - üîπ We extracted the regression **coefficients (b‚ÇÄ, b‚ÇÅ)**.
# - üîπ We **validated** the equation by checking predictions.
# - üîπ We tested the model with **new experience levels**.
# - üîπ We evaluated model accuracy using **MAE and R¬≤ Score**.
# - üîπ **Our regression model is now fully operational! üöÄ**
```

# üìä Map of Regression Error


```python
# üìå Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# üéØ Create a figure and axis for the plot
fig, ax = plt.subplots(figsize=(10, 7))

# üìä Generate a grid of points representing experience (X) and salary (Y)
# Experience range: 0 to 20 years, Salary range: 20,000 to 250,000
X_grid, Y_grid = np.meshgrid(np.linspace(0, 20, 100), np.linspace(20000, 250000, 100))

# ‚öôÔ∏è Compute the regression error at each point
# This represents how far off the predicted salary is from actual salary
Error_grid = np.abs(b0 + b1 * X_grid - Y_grid)

# üé® Plot the filled contour map of the regression error
contour = ax.contourf(X_grid, Y_grid, Error_grid, cmap='coolwarm_r', levels=50)
cbar = plt.colorbar(contour, ax=ax)  # Add color legend (bar)

# üõ† Add contour lines for more details
lines = ax.contour(X_grid, Y_grid, Error_grid, colors='black', linewidths=0.5)
ax.clabel(lines, inline=True, fontsize=8)  # Label contour lines

# üè∑ Set axis labels and title
ax.set_xlabel('Experience (Years)', fontsize=12)  # X-axis: Experience in years
ax.set_ylabel('Salary ($)', fontsize=12)  # Y-axis: Salary in dollars
ax.set_title('üìä Contour Map of Regression Error', fontsize=14, fontweight="bold")  # Main title

# üîé Highlight areas of low and high error
plt.text(2, 50000, "Low Error", fontsize=16, color="blue", fontweight="bold")  # Small error zone
plt.text(15, 200000, "High Error", fontsize=16, color="red", fontweight="bold")  # Large error zone

# üöÄ Show the final plot
plt.show()
```
